### In Synthetic data

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import make_classification
````

````python
X, y = make_classification(n_samples = 250 , n_features = 2 , n_redundant = 0, n_informative = 2, n_clusters_per_class = 1, n_classes = 2, random_state = 3 )

print(X.shape, y.shape)
_, ax = plt.subplots(figsize=( 5 , 4 ))
ax.scatter( X[:, 0] , X[:, 1] , c = y , cmap = 'Paired')
````

````python
# Create an instance
decision_tree = DecisionTreeClassifier()

# Train the model
decision_tree.fit ( X , y ) #(# Samples, # Features)
````

### **Decision boundaries**

````python
# Min-Max values for the grid of decision boundaries
h_min, h_max = X[:,0].min()-0.1 , X[:,0].max()+0.1
v_min, v_max = X[:,1].min()-0.1 , X[:,1].max()+0.1

# Create gridnp.linspace(
h_grid, v_grid = np.meshgrid(np.linspace(h_min,h_max, 200), np.linspace(v_min, v_max, 200))

# print(h_grid.shape, v_grid.shape, h_grid.ravel().shape, v_grid.ravel().shape)
# print(np.c_[h_grid.ravel(), v_grid.ravel()].shape)
# Estimate predictions over the grid
predictions = decision_tree.predict(np.c_[h_grid.ravel(), v_grid.ravel()])
print (predictions.shape)

# Reshape predictions
predictions = predictions.reshape(h_grid.shape)
print(predictions.shape)
````

````python
# Draw grid
_, ax = plt.subplots (figsize=(5,4))
#ax.pcolormesh( h_grid , v_grid , predictions , cmap = "Paired")

# Scatter real values diagrama de dispercion
plot_tree(decision_tree, filled=True, fontsize=5, ax=ax)

#ax.set_xticks(())
#ax.set_yticks(())
````
