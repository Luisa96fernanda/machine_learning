# **Linear Regresion**

```linear_regression``` module documentation [here](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html?highlight=linear%20regression#sklearn.linear_model.LinearRegression)

````python
import numpy as np # Numerical library
import matplotlib.pyplot as plt # plots
from sklearn.linear_model import LinearRegression
%matplotlib inline
````

````python
def f(x) : # Create a function that samples from linear function x son los datos
  y = (0.0918 * x) + 1.2859 + 0.15*np.random.randn (x.shape[0]) # Linear function
  return y #devuelve y ecuacion de la linea y=mx+b(m, pendiente, b es el intervaalo)
  #np.random(es el ruido de la ecuacion lineal)la funcion devuelve datos con factor de ruido

# create training samples
x = np.arange (0, 20, 0.25); print (x.shape)
# Generate samples
y = f(x) ; print (y.shape)

# Scatter plot
_,ax = plt.subplots (figsize = ( 5 , 4 ))
ax.scatter(x, y , color = 'blue')
````

````python
# Create an instance
linear_regression = LinearRegression()

# Train the model
#no tiene numero de caracteriticas X
linear_regression.fit( x.reshape(-1,1) , y ) # (# Samples(muestras), # Features(caracteristicas))
````

````python
# Print estimated values
#linea de pendiente es el .coef
print ("w = " + str(linear_regression.coef_) + ", b = " + str(linear_regression.intercept_))
````

````python
# Create a new sample estimate prediction
new_sample = np.array([5])
print (f(new_sample))

prediction = linear_regression.predict( new_sample.reshape(-1,1) )
print (prediction)
````

````python
# Predict x values
predictions = linear_regression.predict( x.reshape(-1,1) )

# Draw real samples (blue) and predictions (red)
_, ax = plt.subplots (figsize = (5,4))
ax.scatter( x , y , color = 'blue')
ax.plot ( x , predictions , color = 'red')
#ax.scatter ( x , predictions , color = 'red')
````

## In Synthetic data

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
````

````python
X, y = make_regression(n_samples = 100 , n_features = 1, noise = 10, random_state = 5 )

_, ax = plt.subplots (figsize=( 5,4 ))
ax.scatter( X , y , s=10, c = 'b')
print (X.shape, y.shape)
````

````python
linear_regression = LinearRegression()
linear_regression.fit( X.reshape(-1,1) , y )
````

````python
# Imprimir los valores estimados para compararlos respecto a los originales
print ("w = " + str(linear_regression.coef_) + ", b = " + str(linear_regression.intercept_))
````

````python
# # Nueva muestra. Imprime el valor correspondiente de y
# new_sample = np.array([1])

# Predecir la nueva muestra. Imprimir la prediccion
predictions = linear_regression.predict( X.reshape(-1,1))
print (predictions.shape)
````

````python
_,ax = plt.subplots (figsize = ( 5 , 4 ))
ax.scatter( X , y , s=10, color = 'b')
ax.plot( X , predictions , color = 'g', linewidth=1)
````

## In real data

```datasets``` module info [here](https://scikit-learn.org/stable/modules/classes.html#module-sklearn.datasets)

````python
import numpy as np
import matplotlib.pyplot as plt # plots
from sklearn.linear_model import LinearRegression # scikit-learn: libreria de IA para python
from sklearn.datasets import load_diabetes
````

````python
# Cargar la base de datos Diabetes
diabetes_X, diabetes_y = load_diabetes(return_X_y=True)
print(diabetes_X.shape, diabetes_y.shape)
````

````python
# Use only one feature
diabetes_X = load_diabetes()
diabetes_X = diabetes_X.data[:, np.newaxis,2]
print(diabetes_X.shape, diabetes_y.shape)
````

````python
# Linear regression
linear_regression = LinearRegression()

# Train the model
linear_regression.fit( diabetes_X.reshape(-1,1) , diabetes_y )
````

````python
# Make predictions using the testing set
diabetes_y_pred = linear_regression.predict(diabetes_X.reshape(-1,1))


print ("w = " + str( linear_regression.coef_ ) + ", b = " + str( linear_regression.intercept_ ))
````

````python
# Plot outputs
_,ax = plt.subplots (figsize = ( 5 , 4 ))
ax.scatter( diabetes_X , diabetes_y , s=10, color = 'b')
ax.plot( diabetes_X , diabetes_y_pred , color = 'g', linewidth=1)
````
